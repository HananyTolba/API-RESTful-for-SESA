```python
from makeprediction.quasigp import QuasiGPR as qgpr
from makeprediction.kernels import Periodic, RBF, Matern32

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import pandas as pd
import numpy as np
```

```python
kernel = Periodic(period = .4, length_scale = .9) + RBF(length_scale = .5)
kernel =  Matern32(length_scale = .5) + Periodic(period = .3)
```

```python
date  = pd.date_range('2020-01-01 00:00:00',end = '2020-01-04 11:10:00', freq = '5min')
seed = np.random.seed(32)
noisy = .15*np.random.randn(len(date))
data = kernel.simulate(date, seed = seed) + noisy
MEAN, STD = 500, 150
data = MEAN + STD*data
len(date)
```

```python

train_size = int(.7*len(data))

X_train, y_train = date[:train_size], data[:train_size]
X_test, y_test = date[train_size:], data[train_size:]
```

```python
model = qgpr(X_train,y_train, kernel)
model.plotly(X_test, y_test)
```

```python
model.fit()
model.plotly(X_test, y_test)
```

```python
model.predict(X_test)
model.plotly(X_test, y_test)
```

```python
update_data = dict(x_update = X_test, y_update = y_test)
model.update(**update_data,method = 'sliding')
```

```python
model.predict(X_test)
model.plotly(X_test, y_test)
```
