```python
numbers = []
for d in range(2,105):
    if 105 % d == 0:
        numbers.append(d)
```

```python
import numpy as np
row = np.array(numbers).reshape(-1,1)
result = row @ row.T
n,m = np.where(result == 105)
#print('numbers = ', numbers)
#print('possibilite = ', result, sep ='\n')
for i,v in enumerate(n):
    print(f'solution numèro {i+1} est {row[v][0]} x { row[m[i]][0]} = {row[v][0]*row[m[i]][0]}' )
```
